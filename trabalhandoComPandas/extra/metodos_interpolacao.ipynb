```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
```

```python
s = pd.Series(data)
```

```python
s
```

```python
s.fillna(0)
```

```python
s.fillna(method='ffill') #ele copia o valor anterir e completa o campo vazio. cima para baixo
```

```python
s.fillna(method='bfill') #faz que o ffill faz mas ao contrário debaixo pra cima
```

```python
s.fillna(s.mean() )
```

```python
s.fillna(method='ffill', limit = 1)
```
