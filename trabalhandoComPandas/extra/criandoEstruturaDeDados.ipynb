```python
import pandas as pd
```

## Series

```python
data  = [1,2,3,4,5]
```

```python
s = pd.Series(data)
```

```python
s
```

```python
index = ['Linha' + str(i) for i in range(5)]
```

```python
index
```

```python
s = pd.Series(data = data, index = index)
```

```python
s
```

```python
data = {
    'Linha' + str(i) : i + 1 for i in range(5)
}
```

```python
data
```

```python
s = pd.Series(data)
s
```

```python
s1 = s + 2 #soma 2 em cada uma das linhas da serie
s1
```

```python
s2 = s + s1 #soma entre series do mesmo shape
s2
```

##DataFrame

```python
data = [[1,2,3], [4,5,6], [7,8,9]]
data
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
linhas = ['linha' + str(i) for i in range(3)]
linhas
```

```python
df1 = pd.DataFrame(data,linhas)
df1
```

```python
coluna = ['Coluna' + str(i) for i in range(3)]
coluna
```

```python
df1 = pd.DataFrame(data,linhas,coluna)
df1
```

```python
data = {
    'coluna0': {'linha0':1, 'linha1':2, 'linha3':3},
    'coluna1': {'linha0':4, 'linha1':5, 'linha3':6},
    'coluna2': {'linha0':7, 'linha1':8, 'linha3':9},
}
data
```

```python
df2 = pd.DataFrame(data)
df2
```

```python
df3 = pd.DataFrame(data)
df3
```

```python
df4 = pd.concat([df1,df2,df3])
df4
```
