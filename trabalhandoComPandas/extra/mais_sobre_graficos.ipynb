```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('figure', figsize = (15,8))
dados = pd.read_csv('aluguel_residencial_sem_outliers.csv', sep = ";")
dados.head(10)
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
g1.scatter(dados.Valor, dados['Area'], color = 'blue')
g1.set_title('Valor x Área')
g1.set_xlim(left =-400 ,right=30800)
g1.set_ylim(bottom=-200, top=3300)
```

```python
g2.hist(dados.Valor)
g2.set_title('Histograma Valor')
```

```python
dados_g3 = dados['Valor'].sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Sample G3')
```

```python
grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valor = grupo.mean().values
g4.bar(label,valor)
g4.set_title('Valor Médio Por Tipo')
```
