## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep = ";")
```

```python
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = list(['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila'
])
```

```python
residencial
```

```python
dados.head(10)
```

```python
selecao = dados["Tipo"].isin(residencial)
selecao
```

```python
dados_residencial = dados[selecao]
dados_residencial
```

```python
list(dados_residencial["Tipo"].drop_duplicates())
```

```python
dados_residencial.shape
```

```python
eliminacoes = dados.shape[0] - dados_residencial.shape[0]
eliminacoes
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

## Exportar dataFrame para CSV


```python
dados_residencial.to_csv('aluguel_residencial.csv', sep=";")
```
