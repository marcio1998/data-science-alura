```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep=";")
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados["Valor"] + dados["Condominio"] + dados["IPTU"]
```

```python
dados.head(10)
```

```python
dados.columns = ["Erro", "Tipo", "Bairro", "Quartos", "Vagas", "Suites", "Area", "Valor", "Condomínio", "IPTU", "Valor Bruto"]
```

```python
dados.head(10)
```

```python
dados.drop(columns=["Erro"], axis = 1, inplace=True)
```

```python
dados["Valor m2"] = dados["Valor"] / dados["Area"]
```

```python
dados.head(10)
```

```python
dados["Valor Bruto m2"] = (dados["Valor"] / dados["Area"]).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados["Tipo"].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```
