## Seleções de Frequências



```python
import pandas as pd
```

```python
dados = pd.read_csv("aluguel_residencial.csv", sep = ";")
```

```python
dados.head(10)
```

Selecione some os imóveis classificados como apartamento

```python
selecao = dados["Tipo"] == "Apartamento"
selecao
```

```python
n1 = dados[selecao].shape[0]
n1
```

imoveis que são classificados como casa

```python
selecao = (dados["Tipo"] == "Casa") | (dados["Tipo"] == "Casa de Condomínio") |( dados["Tipo"] == "Casa de Vila")
selecao
```

```python
n2 = dados[selecao].shape[0]
n2
```

Selecione os imoveis que a area está entre 60 e 100 metros quadrados

```python
selecao = (dados["Area"] >= 60) & (dados["Area"] <= 100)
selecao
```

```python
n3 = dados[selecao].shape[0]
n3
```

selecione imoveis que tenham pelo menos 4 quartos ou mais e alguel menor que 2000 reais

```python
selecao = (dados["Quartos"] >= 4) & (dados["Valor"] < 2000)
selecao
```

```python
n4 = dados[selecao].shape[0]
n4
```
